Run this to adjust the ninja temp country by country using the model parameters from ninja_temp_adjustment.csv

```python
import os
import pandas as pd
import numpy as np
```

```python
cwd=os.getcwd()
path_raw_ninja_data=os.path.join(cwd, "../../raw_data/ninja_weather/")
path_adjustment_model=os.path.join(cwd, 'ninja_temp_adjustment.csv')

df_adjustment=pd.read_csv(path_adjustment_model, index_col=0)
df_adjustment=df_adjustment[['slope', 'intercept']]
df_adjustment
```

```python
for country in df_adjustment.index:
    
    alpha=df_adjustment.loc[country, 'slope']
    beta=df_adjustment.loc[country, 'intercept']

    df_raw=pd.read_csv(os.path.join(path_raw_ninja_data, 
                        'ninja_weather_country_%s_merra-2_population_weighted.csv' %country),
                        skiprows=2, usecols=[0,2])
    df_raw['time']=pd.to_datetime(df_raw['time'])
    df_raw=df_raw.set_index('time')
    
    df_raw['temperature']=df_raw['temperature']*alpha + beta
    
    df_raw.to_csv('adjusted_ninja_temperature/adjusted_ninja_temp_%s.csv' %country)
    
    print(country)
```
